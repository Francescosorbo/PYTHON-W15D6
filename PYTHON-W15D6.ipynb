```python
import numpy as np
years = 5
a = pd.DataFrame({"Mese": list("GFMAMGLASOND"*years),
                "Anno": np.repeat(list(range(years)), 12),
                "Guadagni": np.random.randint(800, 5000, 12*years)})

anni = a.Anno.unique()
for anno in anni:
    filtro_anno = a.Anno == anno
    print("Anno:", anno)
    somma_cumulativa_annuale = a.loc[filtro_anno, "Guadagni"].cumsum()
    print(somma_cumulativa_annuale)
    print("-"*10)
```
